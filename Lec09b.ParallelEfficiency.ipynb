## Parallel Efficiency

Companion concept to speedup.
* Efficiency; $E = S(n)/n = T(1)/nT(n)$
* Informally: fraction of possible performance realized

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

resources = range(1,65)
p = [0,.5,.9,.95,.99,.999, 1.0]

data_arr = np.empty((len(resources), len(p)))
    
# initialize array using numpy indexing
for i,j in np.ndindex(data_arr.shape):
    # efficiency is s\n
    data_arr[i,j] = (1/(1-p[j]+p[j]/resources[i]))/resources[i]
    
#plot in linear linear space
df = pd.DataFrame(data=data_arr, index=resources, columns=p)
df.plot()
plt.title("linear/linear")
```

```python
%matplotlib inline

#plot in linear linear space
df.plot()
plt.title("log/linear")
plt.semilogx()
```

No common convention about plotting efficiency
* Typical to have efficiency be linear scale and x-axis be either log of linear.
* linear/linear has better intuition
* log/linear is not monotonic
* efficiency varies from 0-1.0 so not log/log available
The shapes of efficiency plots are characteristic. Be careful to read axes.

EFficiency is a more natural notion that speedup. Easy to comprehend. Often less informative because it's indirect.
  * Speedup is the goal! 
  * Efficiency is a measure of how well a program uses parallel computing resources.
